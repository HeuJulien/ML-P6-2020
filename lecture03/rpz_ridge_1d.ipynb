# 1D

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

def rpz_ridge_1d(ols_zero=3,ols_coeff=1,lmbd=0.5):
    x = np.linspace(-1,5,500)
    l2_norm = lambda x, lmbd : lmbd*(x**2)
    ols = lambda x, ols_zero, ols_coeff : ols_coeff*((x-ols_zero)**2)
    
    min_ols = np.argmin(ols(x,ols_zero,ols_coeff)+l2_norm(x,lmbd))
    
    plt.figure()
    plt.plot(x,ols(x,ols_zero,ols_coeff),label='OLS')
    plt.plot(x,l2_norm(x,lmbd),label='lambda * l2_norm^{2}')
    plt.plot(x,ols(x,ols_zero,ols_coeff)+l2_norm(x,lmbd),label='Ridge')
    plt.plot(x[min_ols],ols(x[min_ols],ols_zero,ols_coeff)+l2_norm(x[min_ols],lmbd),'*',markersize=10)
    plt.ylim([-1,15])
    plt.legend()
    plt.title('Lambda = {}'.format(lmbd))
    plt.xlabel('beta_1')
    plt.ylabel('Risk')
    
rpz_ridge_1d()
rpz_ridge_1d(lmbd=3)
```

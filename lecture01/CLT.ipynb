```python
import numpy as np
import matplotlib.pyplot as plt


def tirage(N_tirage=10_000, N_box=100, N_faces = 6):
    """
    Function used to visualize the central limit theorem (CLT)
    with the used of a N_faces-faced dice
    """
    # We throw N_tirage * N_box dices and record their results
    tirages = np.random.randint(1,N_faces+1,size=int(N_tirage*N_box)).reshape((N_box,N_tirage))
    # We compute the mean for each "box"
    means = tirages.mean(axis=1)
    # We plot the distribution of the averages
    plt.hist(means,density=True)
    
    # Simple gaussian function
    gaussian = lambda x, mu, var : 1/np.sqrt(2*np.pi*var) * np.exp(-((x-mu)**2)/(2*var))
    
    # True values of mean/variance of a N_faces-faced dice
    mean_dice = np.arange(1,N_faces+1).mean()
    var_dice = (np.arange(1,N_faces+1)**2).mean() - mean_dice**2

    # Plot of a gaussian pdf with mean and variance following the CLT
    xs = np.linspace(means.min(),means.max(),200)
    plt.plot(xs,gaussian(xs,mean_dice,var_dice/N_tirage),label = N_tirage)
    plt.legend()
```

```python
tirage(10_000,100)
tirage(100_000,100)
```

As we could expect from the central limit theorem, by increasing the number of dices thrown (N_tirage), the variance of the distribution decreases.

Note : Increasing N_box do not change the variance of the distribution, it just allows us to have a smoother distribution

```python
tirage(10_000,1000)
tirage(100_000,1000)
```
